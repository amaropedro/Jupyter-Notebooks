Ativdade de Esteganografia. Esconde uma mensagem tratando caracters como pixels de uma imagem.

```python
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
```

## Com um mensagem pequena

```python
a ="Minha senha é: 14350"
print(a)
```

```python
b = []

for i in a:
  b.append(ord(i))

print(b)
len(b)
```

```python
c = np.uint8(b)
c
```

```python
#size = (altura, largura) 
#de tal modo que altura*largura == comprimento da string de entrada
size = (4, 5) 
d = np.reshape(c, size)
print(d)
```

```python
im = Image.fromarray(d)
im.save("coolimage.bmp")
```
