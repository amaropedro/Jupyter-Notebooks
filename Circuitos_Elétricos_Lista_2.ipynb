```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
t = np.linspace(0, 2, 100)

def func(t):
  return (0.1*math.sin(11177.54*t) + 3*math.cos(11177.54*t)) 

f2 = np.vectorize(func)

fig, ax = plt.subplots()
ax.plot(t, f2(t))
plt.title('i (t)')
ax.axhline(y=0, color='k', linewidth = 0.5)
ax.axvline(x=0, color='k', linewidth = 0.5)
plt.ylabel('i (A)')
plt.xlabel('t (s)')
fig.show()
```

```python
t = np.linspace(0,2,100)

def func(t):
  return 12 - (12*math.cos(11177.54*t) - 268.4*math.sin(11177.54*t)) 

f2 = np.vectorize(func)

fig, ax = plt.subplots()
ax.plot(t, f2(t))
plt.title('Vc (t)')
ax.axhline(y=0, color='k', linewidth = 0.5)
ax.axvline(x=0, color='k', linewidth = 0.5)
plt.ylabel('Vc (V)')
plt.xlabel('t (s)')
fig.show()
```
